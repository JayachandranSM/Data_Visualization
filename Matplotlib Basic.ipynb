```python
import matplotlib.pyplot as plt
import numpy as np
```

## Simple Plot

```python
#plt.figure(figsize=5,5)
plt.plot([1, 2, 3, 4], [1, 4, 2, 3])
```

### Plot with labels

```python
plt.plot([1, 2, 3, 4], [1, 4, 2, 3],marker='o',color='red',linestyle='dashed',
...      linewidth=2, markersize=12)
#plt.plot([1, 2, 3, 4], [1, 2, 3, 4],marker='o',color='BLUE',linestyle='dashed',
...      #linewidth=2, markersize=12)
plt.grid(True)#default
plt.grid(color='b', ls = '-.', lw = 1)
plt.title("Line plot",loc="right")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.xlim([0.1,4])
plt.ylim([0.5,6])
plt.text(1.5,3,'Line plot',rotation = 35, 
               rotation_mode ='anchor')
```

## Subplot

```python
fig = plt.figure()  # an empty figure with no Axes
fig, ax = plt.subplots()  # a figure with a single Axes
fig, ax = plt.subplots(2,2)  # a figure with a 2x2 grid of Axes
```

### Advanced plot

```python
fig = plt.figure()  # an empty figure with no Axes
fig, ax = plt.subplots()  # a figure with a single Axes
plt.plot([1, 2, 3, 4], [1, 4, 2, 3])
ax.spines['bottom'].set_color('blue')
ax.spines['left'].set_color('red')
ax.spines['left'].set_linewidth(2)
ax.spines['right'].set_color(None)
ax.spines['top'].set_color(None)
```

```python
fig, ax = plt.subplots(2, 2)
ax[0][0].plot([1, 2, 3, 4], [1, 4, 2, 3])
ax[0][1].plot([1, 2, 3, 4], [1,2, 3,4])
ax[1][0].plot([1, 2, 3, 4], [1,2, 3,4])
ax[1][1].plot([1, 2, 3, 4], [1,2, 3,4])
```

## Pie Chart

## Mesh Grid

```python
xlist = np.linspace(-3.0, 3.0, 100)
ylist = np.linspace(-3.0, 3.0, 100)
X, Y = np.meshgrid(xlist, ylist)
Z = np.sqrt(X**2 + Y**2)
cp=plt.contourf(X, Y, Z)
plt.grid(True)#default
plt.grid(color='b', ls = '-.', lw = 1)
plt.title("mesh plot",loc="right")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.colorbar(cp)
plt.show()
```

```python
from mpl_toolkits import mplot3d
fig = plt.figure()
ax = plt.axes(projection='3d')
xlist = np.linspace(-3.0, 3.0, 100)
ylist = np.linspace(-3.0, 3.0, 100)
X, Y = np.meshgrid(xlist, ylist)
Z = np.sqrt(X**2 + Y**2)
cp=plt.contourf(X, Y, Z)
#ax.contour3D(X, Y, Z, 50, cmap='gray')
#ax.plot_wireframe(X, Y, Z, color='black')
ax.plot_surface(X,Y,Z,cmap='viridis', edgecolor='none')
plt.show()
```

# Contour

```python
xlist =[4,2]
ylist = [4,2]
X, Y = np.meshgrid(xlist, ylist)

Z=X*Y
cp=plt.contourf(X, Y,Z)
plt.grid(True)#default
plt.grid(color='b', ls = '-.', lw = 1)
plt.title("Line plot",loc="right")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.colorbar(cp)
plt.show()
```

## Gradient

```python
x,y = np.meshgrid(np.arange(-2, 2, .2), np.arange(-2, 2, .25))
z = x*np.exp(-x**2 - y**2)
v, u = np.gradient(z, .2, .2)
plt.subplots()
plt.quiver(x,y,u,v)
plt.show()
```

# Box plot

```python
np.random.seed(10)
collectn_1 = np.random.normal(100, 10, 200)
collectn_2 = np.random.normal(80, 30, 200)
collectn_3 = np.random.normal(90, 20, 200)
collectn_4 = np.random.normal(70, 25, 200)
data_to_plot=[collectn_1,collectn_2,collectn_3,collectn_4]
plt.boxplot(data_to_plot)
plt.show()
```

## Violin plot

```python
np.random.seed(10)
collectn_1 = np.random.normal(100, 10, 200)
collectn_2 = np.random.normal(80, 30, 200)
collectn_3 = np.random.normal(90, 20, 200)
collectn_4 = np.random.normal(70, 25, 200)
data_to_plot=[collectn_1,collectn_2,collectn_3,collectn_4]
plt.violinplot(data_to_plot)
plt.show()
```

# 3D

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
fig = plt.figure()
ax = plt.axes(projection='3d')
z = np.linspace(0, 1, 100)
x = z * np.sin(20 * z)
y = z * np.cos(20 * z)
ax.plot3D(x, y, z, 'gray')
ax.set_title('3D line plot')
plt.show()
```
