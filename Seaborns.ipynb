# Univarite Analysis

```python
import seaborn as sb
import matplotlib.pyplot as plt  # Importing matplotlib.pyplot for figure manipulation

# Assuming you have a 'dataset' DataFrame with a 'ssc_p' column

# Create a figure and set its size
plt.figure(figsize=(5, 3))

# Create the distribution plot using Seaborn
sb.distplot(dataset['ssc_p'],hist=True, kde=True, kde_kws={'color': 'red'})

# Display the plot
plt.show()
```

# Bivarite Analysis : Distribution Plot

```python
import seaborn as sb
import matplotlib.pyplot as plt

# Assuming you have a 'dataset' DataFrame with 'ssc_p' and 'salary' columns
y_axis_range = (200000, 945000)

plt.figure(figsize=(5, 3))
# Create the joint plot with y-axis limits
sb.jointplot(x='ssc_p', y='salary', data=dataset, ylim=y_axis_range)

plt.show()
```
